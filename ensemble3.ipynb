```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pprint
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense
from tensorflow.keras.callbacks import EarlyStopping
import random

# ---------------------------
# Load Dataset
# ---------------------------
def load_dataset():
    data = pd.read_csv('final-dataset.csv')
    X = data.drop(columns=['Signal', 'Label', 'Hurst Component'])
    y = data['Label']
    return X, y

# ---------------------------
# Feature Selection
# ---------------------------
def select_features(X, y, method='f_classif', k=20):
    feature_names = X.columns
    X_scaled = StandardScaler().fit_transform(X)
    selector = SelectKBest(score_func=f_classif, k=k)
    selector.fit(X_scaled, y)
    selected_feature_names = feature_names[selector.get_support()]
    X_selected = selector.transform(X_scaled)
    return X_selected, selected_feature_names

# ---------------------------
# Build DL Models
# ---------------------------
def build_rnn_model(input_shape, units=128):
    model = Sequential()
    model.add(SimpleRNN(units, input_shape=input_shape))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def build_lstm_model(input_shape, units=128):
    model = Sequential()
    model.add(LSTM(units, input_shape=input_shape))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def build_gru_model(input_shape, units=128):
    model = Sequential()
    model.add(GRU(units, input_shape=input_shape))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ---------------------------
# Train DL models
# ---------------------------
def train_dl_model(build_fn, X_train, y_train, epochs=50, batch_size=32):
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    model = build_fn(input_shape=(1, X_train.shape[2]))
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2, callbacks=[early_stop], verbose=0)
    return model

# ---------------------------
# Generate combos of models
# ---------------------------
def generate_combos(models, combo_size, n_combos=7):
    combos = set()
    while len(combos) < n_combos:
        combo = tuple(sorted(random.sample(models, combo_size)))
        combos.add(combo)
    return list(combos)

def combos_to_dict(combos_list, prefix):
    return {f"{prefix}-{i+1}": list(c) for i, c in enumerate(combos_list)}

# ---------------------------
# Evaluate single model
# ---------------------------
def evaluate_single_model(name, y_test, y_pred, probs):
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    spec = tn / (tn + fp)

    fpr, tpr, _ = roc_curve(y_test, probs)
    roc_auc = auc(fpr, tpr)

    print(f"\n{name} Performance:")
    print(f" Accuracy    : {acc:.4f}")
    print(f" Precision   : {prec:.4f}")
    print(f" Recall      : {rec:.4f}")
    print(f" F1 Score    : {f1:.4f}")
    print(f" Specificity : {spec:.4f}")
    print(f" ROC AUC     : {roc_auc:.4f}")

    plt.figure(figsize=(6,6))
    plt.plot(fpr, tpr, label=f'ROC AUC = {roc_auc:.2f}')
    plt.plot([0,1], [0,1], linestyle='--', color='gray')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"{name} ROC Curve")
    plt.legend(loc="lower right")
    plt.show()

# ---------------------------
# Ensemble Evaluation
# ---------------------------
def evaluate_ensemble(name, y_test, preds_list, probs_list):
    # Majority voting (vote=1 if sum preds >= half)
    ensemble_pred = (np.sum(preds_list, axis=0) >= (len(preds_list) / 2)).astype(int)
    
    acc = accuracy_score(y_test, ensemble_pred)
    prec = precision_score(y_test, ensemble_pred)
    rec = recall_score(y_test, ensemble_pred)
    f1 = f1_score(y_test, ensemble_pred)
    tn, fp, fn, tp = confusion_matrix(y_test, ensemble_pred).ravel()
    spec = tn / (tn + fp)
    
    avg_probs = np.mean(probs_list, axis=0)
    fpr, tpr, _ = roc_curve(y_test, avg_probs)
    roc_auc = auc(fpr, tpr)

    print(f"\n{name} Majority Voting Ensemble Performance:")
    print(f" Accuracy    : {acc:.4f}")
    print(f" Precision   : {prec:.4f}")
    print(f" Recall      : {rec:.4f}")
    print(f" F1 Score    : {f1:.4f}")
    print(f" Specificity : {spec:.4f}")
    print(f" ROC AUC     : {roc_auc:.4f}")

    plt.figure(figsize=(6,6))
    plt.plot(fpr, tpr, label=f'ROC AUC = {roc_auc:.2f}')
    plt.plot([0,1], [0,1], linestyle='--', color='gray')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"{name} Ensemble ROC Curve")
    plt.legend(loc="lower right")
    plt.show()

    return ensemble_pred

# ---------------------------
# Main script
# ---------------------------
X, y = load_dataset()
X_selected, selected_features = select_features(X, y, method='f_classif', k=20)

# Reshape for DL models (samples, timesteps=1, features)
X_selected_dl = X_selected.reshape((X_selected.shape[0], 1, X_selected.shape[1]))

# Split for training/testing
X_train_dl, X_test_dl, y_train, y_test = train_test_split(X_selected_dl, y, test_size=0.2, stratify=y, random_state=42)
X_train_ml, X_test_ml = train_test_split(X_selected, test_size=0.2, stratify=y, random_state=42)  # ML models take 2D input

# Train DL models
model_rnn = train_dl_model(build_rnn_model, X_train_dl, y_train)
model_lstm = train_dl_model(build_lstm_model, X_train_dl, y_train)
model_gru = train_dl_model(build_gru_model, X_train_dl, y_train)

# Predict DL models
probs_rnn = model_rnn.predict(X_test_dl).ravel()
y_pred_rnn = (probs_rnn > 0.5).astype(int)

probs_lstm = model_lstm.predict(X_test_dl).ravel()
y_pred_lstm = (probs_lstm > 0.5).astype(int)

probs_gru = model_gru.predict(X_test_dl).ravel()
y_pred_gru = (probs_gru > 0.5).astype(int)

# Train ML models
model_svm = SVC(probability=True, random_state=42)
model_svm.fit(X_train_ml, y_train)
probs_svm = model_svm.predict_proba(X_test_ml)[:, 1]
y_pred_svm = model_svm.predict(X_test_ml)

model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(X_train_ml, y_train)
probs_dt = model_dt.predict_proba(X_test_ml)[:, 1]
y_pred_dt = model_dt.predict(X_test_ml)

model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train_ml, y_train)
probs_rf = model_rf.predict_proba(X_test_ml)[:, 1]
y_pred_rf = model_rf.predict(X_test_ml)

model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_lr.fit(X_train_ml, y_train)
probs_lr = model_lr.predict_proba(X_test_ml)[:, 1]
y_pred_lr = model_lr.predict(X_test_ml)

model_knn = KNeighborsClassifier()
model_knn.fit(X_train_ml, y_train)
probs_knn = model_knn.predict_proba(X_test_ml)[:, 1]
y_pred_knn = model_knn.predict(X_test_ml)

# Store predictions and probabilities
all_preds = {
    'RNN': y_pred_rnn,
    'LSTM': y_pred_lstm,
    'GRU': y_pred_gru,
    'SVM': y_pred_svm,
    'DecisionTree': y_pred_dt,
    'RandomForest': y_pred_rf,
    'LogisticRegression': y_pred_lr,
    'KNN': y_pred_knn
}

all_probs = {
    'RNN': probs_rnn,
    'LSTM': probs_lstm,
    'GRU': probs_gru,
    'SVM': probs_svm,
    'DecisionTree': probs_dt,
    'RandomForest': probs_rf,
    'LogisticRegression': probs_lr,
    'KNN': probs_knn
}

all_models = list(all_preds.keys())

# Generate combos
combos_3 = generate_combos(all_models, 3, 7)
combos_5 = generate_combos(all_models, 5, 7)
combos_7 = generate_combos(all_models, 7, 7)

combos_dict_3 = combos_to_dict(combos_3, "Combo 3")
combos_dict_5 = combos_to_dict(combos_5, "Combo 5")
combos_dict_7 = combos_to_dict(combos_7, "Combo 7")

# ===== Evaluate single models =====
print("===== Single Model Performances =====")
for model_name in all_models:
    evaluate_single_model(model_name, y_test, all_preds[model_name], all_probs[model_name])

# ===== Evaluate 3-of-7 combos =====
print("\n===== Ensembles: 3-of-7 Combos =====")
for combo_name, model_names in combos_dict_3.items():
    print(f"\nEvaluating Ensemble: {combo_name} with models: {model_names}")
    preds_list = [all_preds[m] for m in model_names]
    probs_list = [all_probs[m] for m in model_names]
    evaluate_ensemble(combo_name, y_test, preds_list, probs_list)

# ===== Evaluate 5-of-7 combos =====
print("\n===== Ensembles: 5-of-7 Combos =====")
for combo_name, model_names in combos_dict_5.items():
    print(f"\nEvaluating Ensemble: {combo_name} with models: {model_names}")
    preds_list = [all_preds[m] for m in model_names]
    probs_list = [all_probs[m] for m in model_names]
    evaluate_ensemble(combo_name, y_test, preds_list, probs_list)

# ===== Evaluate 7-of-7 combos =====
print("\n===== Ensembles: 7-of-7 Combos =====")
for combo_name, model_names in combos_dict_7.items():
    print(f"\nEvaluating Ensemble: {combo_name} with models: {model_names}")
    preds_list = [all_preds[m] for m in model_names]
    probs_list = [all_probs[m] for m in model_names]
    evaluate_ensemble(combo_name, y_test, preds_list, probs_list)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Sample data as before (replace with your full data)
data = [
    # Individual Models
    ['RandomForest', 0.8214, 0.7892, 0.8571, 0.7929, 0.8192, 0.9042, 'Individual Model'],
    ['DecisionTree', 0.8214, 0.7805, 0.8643, 0.7857, 0.8205, 0.9080, 'Individual Model'],
    ['KNN', 0.7679, 0.7341, 0.8071, 0.7143, 0.7687, 0.8924, 'Individual Model'],
    ['SVM', 0.7500, 0.7139, 0.7857, 0.7071, 0.7483, 0.8883, 'Individual Model'],
    ['LogisticRegression', 0.7143, 0.6825, 0.7500, 0.6857, 0.7156, 0.8753, 'Individual Model'],
    ['GRU', 0.7500, 0.7315, 0.7714, 0.7357, 0.7513, 0.8859, 'Individual Model'],
    ['LSTM', 0.7393, 0.7002, 0.7857, 0.7000, 0.7413, 0.8685, 'Individual Model'],
    ['RNN', 0.7250, 0.6992, 0.7500, 0.7000, 0.7222, 0.8606, 'Individual Model'],

    # Combo 3 (just few samples here)
    ['Combo 3-1 DecisionTree, GRU, RandomForest', 0.8679, 0.8552, 0.8857, 0.8500, 0.8702, 0.9245, '3-Model Combo'],
    ['Combo 3-2 KNN, RandomForest, SVM', 0.8000, 0.7593, 0.8786, 0.7214, 0.8146, 0.9116, '3-Model Combo'],
    ['Combo 3-3 DecisionTree, RandomForest, SVM', 0.8786, 0.8630, 0.9000, 0.8571, 0.8811, 0.9267, '3-Model Combo'],

    # Combo 5 (few samples)
    ['Combo 5-1 DecisionTree, GRU, KNN, LogisticRegression, SVM', 0.7393, 0.7134, 0.8000, 0.6786, 0.7542, 0.8890, '5-Model Combo'],
    ['Combo 5-2 GRU, KNN, LSTM, LogisticRegression, SVM', 0.7179, 0.6993, 0.7643, 0.6714, 0.7304, 0.8411, '5-Model Combo'],

    # Combo 7 (few samples)
    ['Combo 7-1 GRU, KNN, LSTM, LogisticRegression, RNN, RandomForest, SVM', 0.7286, 0.7105, 0.7714, 0.6857, 0.7397, 0.8764, '7-Model Combo'],
    ['Combo 7-2 DecisionTree, GRU, KNN, LSTM, LogisticRegression, RandomForest, SVM', 0.7607, 0.7417, 0.8000, 0.7214, 0.7698, 0.9070, '7-Model Combo']
]

df = pd.DataFrame(data, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'Specificity', 'F1-Score', 'ROC AUC', 'Type'])

alias_map = {
    'RandomForest': 'RF',
    'DecisionTree': 'DT',
    'KNN': 'KNN',
    'SVM': 'SVM',
    'LogisticRegression': 'LR',
    'GRU': 'GRU',
    'LSTM': 'LSTM',
    'RNN': 'RNN'
}

def replace_names_with_alias(combo_name):
    models_part = re.sub(r'^Combo\s*\d+-\d+\s*', '', combo_name).strip()
    models_list = [m.strip() for m in models_part.split(',')]
    aliases = [alias_map.get(m, m) for m in models_list]
    return ", ".join(aliases)

df['Model_Alias'] = df['Model'].apply(replace_names_with_alias)

import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
metrics = ['Accuracy', 'Precision', 'Recall', 'Specificity', 'F1-Score', 'ROC AUC']

def plot_metrics_with_values(df, model_type):
    df_sub = df[df['Type'] == model_type].copy()
    df_sub = df_sub.sort_values('Accuracy', ascending=False)

    plt.figure(figsize=(16, 9))
    
    for i, metric in enumerate(metrics, 1):
        plt.subplot(2, 3, i)
        ax = sns.barplot(x=metric, y='Model_Alias', data=df_sub, palette='viridis')

        plt.title(f'{metric} - {model_type}')
        plt.xlim(0, 1)
        plt.xlabel(metric)
        plt.ylabel('Model Combination')

        # Add numeric labels to bars
        for p in ax.patches:
            width = p.get_width()
            y = p.get_y() + p.get_height() / 2
            ax.text(width + 0.01, y, f'{width:.4f}', va='center', fontsize=8)
    
    plt.tight_layout(rect=[0, 0.15, 1, 1])
    combos_aliases = df_sub['Model_Alias'].tolist()
    caption_text = f"{model_type} Models/Combinations (Aliases): " + "; ".join(combos_aliases)
    plt.figtext(0.5, 0.05, caption_text, wrap=True, ha='center', fontsize=9)
    plt.show()

# Example plotting call
plot_metrics_with_values(df, 'Individual Model')
plot_metrics_with_values(df, '3-Model Combo')
plot_metrics_with_values(df, '5-Model Combo')
plot_metrics_with_values(df, '7-Model Combo')
```
