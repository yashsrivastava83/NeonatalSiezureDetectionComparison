```python
def generate_summary(df_top):
    best_acc = df_top.iloc[0]
    best_f1 = df_top.sort_values('F1-Score', ascending=False).iloc[0]
    best_auc = df_top.sort_values('ROC AUC', ascending=False).iloc[0]

    print("Summary of Top Models Performance:\n")
    print(f"Best Accuracy: {best_acc['Alias']} with {best_acc['Accuracy']:.4f}")
    print(f"Best F1-Score: {best_f1['Alias']} with {best_f1['F1-Score']:.4f}")
    print(f"Best ROC AUC: {best_auc['Alias']} with {best_auc['ROC AUC']:.4f}\n")

    # Combo vs Individual analysis
    combo_count = df_top['Type'].str.contains('Combo').sum()
    indiv_count = len(df_top) - combo_count
    print(f"Among the top {len(df_top)} models, {combo_count} are combo models and {indiv_count} are individual models.")

    print("\nObservations:")
    if combo_count > indiv_count:
        print("- Combo models tend to outperform individual models in this dataset.")
    else:
        print("- Individual models hold their ground and sometimes outperform combos.")

    # Highlight any significant differences
    accuracy_range = df_top['Accuracy'].max() - df_top['Accuracy'].min()
    if accuracy_range > 0.1:
        print("- Significant variation in accuracy among top models; model choice greatly affects performance.")
    else:
        print("- Accuracy is relatively stable among top models.")

# Use the function on your top 5 dataframe
generate_summary(top5_acc)
```
